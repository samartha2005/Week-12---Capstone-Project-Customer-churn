```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
#Load Clean Data

df = pd.read_csv("D:/customer-churn-capstone/data/processed/churn_cleaned.csv")
df.head()
```

```python
#Churn Distribution Visualization

plt.figure(figsize=(6,4))
sns.countplot(x="Churn", data=df)
plt.title("Churn Distribution")
plt.show()
```

### Insight:
The dataset is highly imbalanced with significantly fewer churn cases.
This indicates the company retains most customers but needs to focus on the small high-risk segment.

```python
#Numerical Feature Analysis

#Tenure vs Churn

plt.figure(figsize=(8,5))
sns.boxplot(x="Churn", y="Tenure", data=df)
plt.title("Tenure vs Churn")
plt.show()
```

### Insight:
Customers with lower tenure are more likely to churn.
New customers are at higher risk.

```python
#MonthlyCharges vs Churn

plt.figure(figsize=(8,5))
sns.boxplot(x="Churn", y="MonthlyCharges", data=df)
plt.title("Monthly Charges vs Churn")
plt.show()
```

### Insight:
Customers with higher monthly charges show higher churn tendency.
Pricing strategy may influence churn behavior.

```python
#TotalCharges vs Churn

plt.figure(figsize=(8,5))
sns.boxplot(x="Churn", y="TotalCharges", data=df)
plt.title("Total Charges vs Churn")
plt.show()
```

### Insight:
Customers with lower total charges tend to churn more, indicating early-stage churn behavior.

```python
#Categorical Feature Analysis

#Contract Type vs Churn

plt.figure(figsize=(8,5))
sns.countplot(x="Contract", hue="Churn", data=df)
plt.title("Contract Type vs Churn")
plt.xticks(rotation=45)
plt.show()
```

### Insight:
Month-to-month contracts have significantly higher churn.
Long-term contracts reduce churn risk.

```python
#Payment Method vs Churn

plt.figure(figsize=(8,5))
sns.countplot(x="PaymentMethod", hue="Churn", data=df)
plt.title("Payment Method vs Churn")
plt.xticks(rotation=45)
plt.show()
```

### Insight:
Certain payment methods may correlate with higher churn.
Further analysis required for business interpretation.

```python
#Paperless Billing vs Churn

plt.figure(figsize=(6,4))
sns.countplot(x="PaperlessBilling", hue="Churn", data=df)
plt.title("Paperless Billing vs Churn")
plt.show()
```
